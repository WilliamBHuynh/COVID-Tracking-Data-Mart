# Date dimensions

```python
dates = pd.read_csv("Date_Table.csv")
```

```python
dates.head()
```

```python
Onset_dimension = pd.DataFrame(columns=['Onset_Date_Surrogate_Key','Onset_Date_Key', 'Day','Month', 'Year', 'Day_of_Week',
                                     'Week_of_Year', 'Weekend', 'Holiday', 'Season', 'Date_Full_Format'])
Reported_dimension = pd.DataFrame(columns=['Reported_Date_Surrogate_Key','Reported_Date_Key', 'Day','Month', 'Year', 'Day_of_Week',
                                     'Week_of_Year', 'Weekend', 'Holiday', 'Season', 'Date_Full_Format'])
Test_dimension = pd.DataFrame(columns=['Test_Date_Surrogate_Key','Test_Date_Key', 'Day','Month', 'Year', 'Day_of_Week',
                                     'Week_of_Year', 'Weekend', 'Holiday', 'Season', 'Date_Full_Format'])
Specimen_dimension = pd.DataFrame(columns=['Specimen_Date_Surrogate_Key','Specimen_Date_Key', 'Day','Month', 'Year', 'Day_of_Week',
                                     'Week_of_Year', 'Weekend', 'Holiday', 'Season', 'Date_Full_Format'])
```

```python
#Onset_dimension 

for idx, row in dates.iterrows():
    onset_row = [idx + 100]
    onset_row.append(row["Date_Key"])
    onset_row.append(row['Day'])
    onset_row.append(row['Month'])
    onset_row.append(row['Year'])
    onset_row.append(row['Day_of_Week'])  
    onset_row.append(row["Week_in_Year"])
    onset_row.append(row['Weekend_YN'])
    onset_row.append(row['Holiday_YN'])
    onset_row.append(row['Season'])
    onset_row.append(row['Date_Full_Format']) 
    Onset_dimension.loc[len(Onset_dimension)] = onset_row

Onset_dimension
```

```python
#Reported_dimension 

for idx, row in dates.iterrows():
    reported_row = [idx + 200]
    reported_row.append(row["Date_Key"])
    reported_row.append(row['Day'])
    reported_row.append(row['Month'])
    reported_row.append(row['Year'])
    reported_row.append(row['Day_of_Week'])  
    reported_row.append(row["Week_in_Year"])
    reported_row.append(row['Weekend_YN'])
    reported_row.append(row['Holiday_YN'])
    reported_row.append(row['Season'])
    reported_row.append(row['Date_Full_Format']) 
    Reported_dimension.loc[len(Reported_dimension)] = reported_row

Reported_dimension
```

```python
#Test_dimension 

for idx, row in dates.iterrows():
    test_row = [idx + 300]
    test_row.append(row["Date_Key"])
    test_row.append(row['Day'])
    test_row.append(row['Month'])
    test_row.append(row['Year'])
    test_row.append(row['Day_of_Week'])  
    test_row.append(row["Week_in_Year"])
    test_row.append(row['Weekend_YN'])
    test_row.append(row['Holiday_YN'])
    test_row.append(row['Season'])
    test_row.append(row['Date_Full_Format']) 
    Test_dimension.loc[len(Test_dimension)] = test_row

Test_dimension
```

```python
#Specimen_dimension 

for idx, row in dates.iterrows():
    specimen_row = [idx + 400]
    specimen_row.append(row["Date_Key"])
    specimen_row.append(row['Day'])
    specimen_row.append(row['Month'])
    specimen_row.append(row['Year'])
    specimen_row.append(row['Day_of_Week'])  
    specimen_row.append(row["Week_in_Year"])
    specimen_row.append(row['Weekend_YN'])
    specimen_row.append(row['Holiday_YN'])
    specimen_row.append(row['Season'])
    specimen_row.append(row['Date_Full_Format']) 
    Specimen_dimension.loc[len(Specimen_dimension)] = specimen_row

Specimen_dimension
```

# Patient dimension

```python
cases.head()
```

```python
Patient_dimension = pd.DataFrame(columns=['Patient_Surrogate_Key','Patient_Key', 'Age', 'Gender', 'Acquisition_Group', 
                                          'Outbreak_Related'])
```

```python
age_unique = cases['Age_Group'].unique()
gender_unique = cases['Client_Gender'].unique()
acq_unique = cases['Case_AcquisitionInfo'].unique()
or_unique = cases['Outbreak_Related'].unique()

print(age_unique, "\n", gender_unique, "\n", acq_unique, "\n", or_unique)
```

```python
index = pd.MultiIndex.from_product([age_unique, gender_unique, acq_unique, or_unique], 
                                   names = ["Age", "Gender", "Acquisition_Group", "Outbreak_Related"])

joined = pd.DataFrame(index = index).reset_index()
joined
```

```python
for idx, row in joined.iterrows():
    
    patient_row = [idx + 1000]
    patient_row.append(idx)
    patient_row.append(row['Age'])
    patient_row.append(row['Gender'])
    patient_row.append(row['Acquisition_Group'])
    patient_row.append(row['Outbreak_Related'])  
    Patient_dimension.loc[len(Patient_dimension)] = patient_row

Patient_dimension
```

```python
# export patient dimension to csv file

Patient_dimension.to_csv("Patient_dimension.csv",index=False)
```
