# Date dimensions

```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

```python
dates = pd.read_csv("Extended_Date_Table.csv")
```

```python
dates.head()
```

```python
Onset_dimension = pd.DataFrame(columns=['Onset_Date_Surrogate_Key','Onset_Date_Key', 'Date_Full_Format', 'Day','Month', 
                                        'Year', 'Day_of_Week_Name','Day_of_Week_Num', 'Week_in_Year', 'Weekend', 'Holiday', 
                                        'Season'])

Reported_dimension = pd.DataFrame(columns=['Reported_Date_Surrogate_Key','Reported_Date_Key', 'Date_Full_Format', 'Day','Month',
                                           'Year', 'Day_of_Week_Name','Day_of_Week_Num', 'Week_in_Year', 'Weekend', 'Holiday',
                                           'Season'])

Test_dimension = pd.DataFrame(columns=['Test_Date_Surrogate_Key','Test_Date_Key', 'Date_Full_Format', 'Day','Month', 
                                       'Year', 'Day_of_Week_Name','Day_of_Week_Num', 'Week_in_Year', 'Weekend', 'Holiday', 
                                       'Season'])

Specimen_dimension = pd.DataFrame(columns=['Specimen_Date_Surrogate_Key','Specimen_Date_Key', 'Date_Full_Format', 'Day','Month',
                                           'Year', 'Day_of_Week_Name','Day_of_Week_Num', 'Week_in_Year', 'Weekend', 'Holiday', 
                                           'Season'])
```

```python
#Onset_dimension 

for idx, row in dates.iterrows():
    onset_row = [idx + 100]
    onset_row.append(row["Date_Key"])
    onset_row.append(row['Date_Full_Format']) 
    onset_row.append(row['Day'])
    onset_row.append(row['Month'])
    onset_row.append(row['Year'])
    onset_row.append(row['Day_of_Week_Name'])  
    onset_row.append(row['Day_of_Week_Num']) 
    onset_row.append(row["Week_in_Year"])
    onset_row.append(row['Weekend_YN'])
    onset_row.append(row['Holiday_YN'])
    onset_row.append(row['Season'])
    Onset_dimension.loc[len(Onset_dimension)] = onset_row

Onset_dimension
```

```python
#Reported_dimension 

for idx, row in dates.iterrows():
    reported_row = [idx + 500]
    reported_row.append(row["Date_Key"])
    reported_row.append(row['Date_Full_Format']) 
    reported_row.append(row['Day'])
    reported_row.append(row['Month'])
    reported_row.append(row['Year'])
    reported_row.append(row['Day_of_Week_Name'])  
    reported_row.append(row['Day_of_Week_Num']) 
    reported_row.append(row["Week_in_Year"])
    reported_row.append(row['Weekend_YN'])
    reported_row.append(row['Holiday_YN'])
    reported_row.append(row['Season'])
    Reported_dimension.loc[len(Reported_dimension)] = reported_row

Reported_dimension
```

```python
#Test_dimension 

for idx, row in dates.iterrows():
    test_row = [idx + 1000]
    test_row.append(row["Date_Key"])
    test_row.append(row['Date_Full_Format']) 
    test_row.append(row['Day'])
    test_row.append(row['Month'])
    test_row.append(row['Year'])
    test_row.append(row['Day_of_Week_Name'])  
    test_row.append(row['Day_of_Week_Num']) 
    test_row.append(row["Week_in_Year"])
    test_row.append(row['Weekend_YN'])
    test_row.append(row['Holiday_YN'])
    test_row.append(row['Season'])
    Test_dimension.loc[len(Test_dimension)] = test_row

Test_dimension
```

```python
#Specimen_dimension 

for idx, row in dates.iterrows():
    specimen_row = [idx + 1500]
    specimen_row.append(row["Date_Key"])
    specimen_row.append(row['Date_Full_Format']) 
    specimen_row.append(row['Day'])
    specimen_row.append(row['Month'])
    specimen_row.append(row['Year'])
    specimen_row.append(row['Day_of_Week_Name'])  
    specimen_row.append(row['Day_of_Week_Num']) 
    specimen_row.append(row["Week_in_Year"])
    specimen_row.append(row['Weekend_YN'])
    specimen_row.append(row['Holiday_YN'])
    specimen_row.append(row['Season'])
    Specimen_dimension.loc[len(Specimen_dimension)] = specimen_row

Specimen_dimension
```

```python
# export date dimensions to csv files

Onset_dimension.to_csv("Onset_Date_dimension.csv",index=False)
Reported_dimension.to_csv("Reported_Date_dimension.csv",index=False)
Test_dimension.to_csv("Test_Date_dimension.csv",index=False)
Specimen_dimension.to_csv("Specimen_Date_dimension.csv",index=False)
```
