## PHU dimension 

```python
import pandas as pd
import numpy as np

from datetime import datetime, timedelta
```

```python
cases = pd.read_csv("conposcovidloc.csv")
```

```python
cases.head()
```

```python
PHU_dimension = pd.DataFrame(columns=['PHU_Surrogate_Key','PHU_Key', 'PHU_Name','Address', 'City', 'Province',
                                     'Postal_Code', 'URL', 'Latitude', 'Longitude'])
```

```python
PHU_dimension

for idx, row in cases.iterrows():
    
    if row["Reporting_PHU"] == "Toronto Public Health":
        PHU_row = [idx + 100]
        PHU_row.append(row["Row_ID"])
        PHU_row.append(row['Reporting_PHU'])
        PHU_row.append(row['Reporting_PHU_Address'])
        PHU_row.append(row['Reporting_PHU_City'])
        PHU_row.append("Ontario")  
        PHU_row.append(row['Reporting_PHU_Postal_Code'])
        PHU_row.append(row['Reporting_PHU_Website'])
        PHU_row.append(row['Reporting_PHU_Latitude'])
        PHU_row.append(row['Reporting_PHU_Longitude'])        
        PHU_dimension.loc[len(PHU_dimension)] = PHU_row
        break
for idx, row in cases.iterrows():
    
    if row["Reporting_PHU"] == "Ottawa Public Health":
        PHU_row = [idx + 100]
        PHU_row.append(row["Row_ID"])
        PHU_row.append(row['Reporting_PHU'])
        PHU_row.append(row['Reporting_PHU_Address'])
        PHU_row.append(row['Reporting_PHU_City'])
        PHU_row.append("Ontario")  
        PHU_row.append(row['Reporting_PHU_Postal_Code'])
        PHU_row.append(row['Reporting_PHU_Website'])
        PHU_row.append(row['Reporting_PHU_Latitude'])
        PHU_row.append(row['Reporting_PHU_Longitude'])        
        PHU_dimension.loc[len(PHU_dimension)] = PHU_row
        break
        
```

```python
PHU_dimension
```

```python
# export PHU dimension to csv file

PHU_dimension.to_csv("PHU_dimension.csv",index=False)
```

# Patient dimension

```python
cases.head()
```

```python
Patient_dimension = pd.DataFrame(columns=['Patient_Surrogate_Key','Patient_Key', 'Age', 'Gender', 'Acquisition_Group', 
                                          'Outbreak_Related'])
```

```python
age_unique = cases['Age_Group'].unique()
gender_unique = cases['Client_Gender'].unique()
acq_unique = cases['Case_AcquisitionInfo'].unique()
or_unique = cases['Outbreak_Related'].unique()

print(age_unique, "\n", gender_unique, "\n", acq_unique, "\n", or_unique)
```

```python
index = pd.MultiIndex.from_product([age_unique, gender_unique, acq_unique, or_unique], 
                                   names = ["Age", "Gender", "Acquisition_Group", "Outbreak_Related"])

joined = pd.DataFrame(index = index).reset_index()
joined
```

```python
for idx, row in joined.iterrows():
    
    patient_row = [idx + 1000]
    patient_row.append(idx)
    patient_row.append(row['Age'])
    patient_row.append(row['Gender'])
    patient_row.append(row['Acquisition_Group'])
    patient_row.append(row['Outbreak_Related'])  
    Patient_dimension.loc[len(Patient_dimension)] = patient_row

Patient_dimension
```

```python
for idx, row in Patient_dimension.iterrows():
    if row["Outbreak_Related"] != "Yes":
        row["Outbreak_Related"] = "No"
```

```python
Patient_dimension
```

```python
# export patient dimension to csv file

Patient_dimension.to_csv("Patient_dimension.csv",index=False)
```
